```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn_extra.cluster import KMedoids

seed = 12
```

```python
variables = ["id", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type"]

df = pd.read_csv('glass.data', header=None)
df.columns = variables

# Replace label 7 by 4 and to have only labels from 1 to 6
print(df["Type"].unique())

df["Type"] = df["Type"].replace(7, 4)
df["Type"] = df["Type"].add(-1)

print(df["Type"].unique())

features = df.drop(["Type"], axis=1)
features.head()
```

```python
# Scatter plot of the true types (alone)
groups = df.groupby('Type')

# Plot
fig, ax = plt.subplots(figsize=(16,12))
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='.', linestyle='', ms=20, label=label)
plt.legend()
plt.show()
```

## KMeans clustering 

```python
kmeans = KMeans(n_clusters=6, random_state=seed).fit_predict(features)
df["kmeans"] = kmeans.tolist()
```

```python
groups = df.groupby('Type')

# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='o', linestyle='', ms=5, label=label)
ax.legend()

plt.show()
```

```python
# Scatter plot of the true types
groups = df.groupby('Type')

# Plot
fig, ax = plt.subplots(figsize=(16,12))
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='x', linestyle='', ms=10, label=label)
plt.legend()

# Scatter plot of the kmeans clusters
groups = df.groupby('kmeans')

# Plot
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='+', linestyle='', ms=10, label=label)
plt.legend()

plt.show()
```

## PAM

```python
pam = KMedoids(n_clusters=6, method="pam",random_state=seed).fit_predict(features)
df["pam"] = pam.tolist()
```

```python
# Scatter plot of the true types
groups = df.groupby('Type')

# Plot
fig, ax = plt.subplots(figsize=(16,12))
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='x', linestyle='', ms=10, label=label)
plt.legend()

# Scatter plot of the pam clusters
groups = df.groupby('pam')

# Plot
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='+', linestyle='', ms=10, label=label)
plt.legend()

plt.show()
```

## CAH

```python
cah = AgglomerativeClustering(n_clusters=6).fit_predict(features)
df["cah"] = cah.tolist()
```

```python
# Scatter plot of the true types
groups = df.groupby('Type')

# Plot
fig, ax = plt.subplots(figsize=(16, 12))
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='x', linestyle='', ms=10, label=label)
plt.legend()


# Scatter plot of the cah clusters
groups = df.groupby('cah')

# Plot
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for label, group in groups:
    ax.plot(group.Na, group.Al, marker='+', linestyle='', ms=10, label=label)
plt.legend()

plt.show()
```
