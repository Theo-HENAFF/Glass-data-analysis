## Corelation between features

```python
cor = df.drop(['id'], axis=1).corr()

fig, ax = plt.subplots(figsize=(8,7))         # Sample figsize in inches
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds, ax=ax)

plt.show()
```

The correlation coefficient has values between -1 to 1
- A value closer to 0 implies weaker correlation (exact 0 implying no correlation)

- A value closer to 1/-1 implies stronger positive/negative correlation

Thus, we want a set of values with low correlation together but a strong one with Type

```python
cor_with_Type = abs(cor["Type"])
cor_with_Type.sort_values(ascending=False)
```

As I thought while looking at the pairplot, Mg as a strong correlation to the Type

But now we can easily see that the K and Ca has nearly no influence on the Type

```python
top5_cols = ['Mg', 'Al', 'Ba', 'Na', "Fe"]

top5_cor = df[top5_cols].corr()
sns.heatmap(top5_cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

As we can see, the some features are quite correlated with each other, but no one has more intra-correletion than correlation than with Type
